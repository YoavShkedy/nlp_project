```python
import os
import openai

# Set API key and base URL
# openai.api_key = 
# openai.api_base = 
openai.api_type = "azure"
openai.api_version = "2023-05-15"  # Use the correct API version

# Specify your deployment ID (the name of the deployed model)
deployment_id = "gpt-4"
```

```python
def generate_summary(prompt, text, category):
  new_prompt = f"{prompt} :{text}"
  response = openai.ChatCompletion.create(
    engine=deployment_id,  # Use your correct deployment ID for ChatCompletion API
    messages=[
        {"role": "system", "content": f"You are very{category}"},
        {"role": "user", "content": new_prompt}
    ],
    max_tokens=600
)
  return response['choices'][0]['message']['content'].strip()
```

```python
def evaluate_summary(summary, text):
  prompt = f"For each of the following category evaluate every summary according to the text from 1-5 points. 1 is the worst and 5 is the best. The categories are: 1. Complete - The summarization should have the main points the text includes. 2. Concise - the summary should be concise. 3. Accuracy - The summary should be accurate with no incorrect information. 4. Hallucinations - The summary should relate to the text and not be in a different weird context and having hallucinations. The text is {text}. The summary is {summary}"
  response = openai.ChatCompletion.create(
    engine=deployment_id,  # Use your correct deployment ID for ChatCompletion API
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ],
    max_tokens=600
)

# Extract the response from 'message' instead of 'text'
  return response['choices'][0]['message']['content'].strip()
```
